```python
sample = read_text()

def read_text ():
    rotations = []
    for line in sample:
        line = line.strip()
        direction = line[0]  # First character is 'L' or 'R'
        distance = int(line[1:])  # Rest is the number
        rotations.append((direction, distance))

# Check what we parsed
print("First few rotations:", rotations[:5])
```
