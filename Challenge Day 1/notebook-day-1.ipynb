```python
# --- Day 1: Secret Entrance ---

with open ('sample.txt', 'r') as sample:
    rotations = []
    for line in sample:
        line = line.strip()
        direction = line[0] 
        distance = int(line[1:])  
        rotations.append((direction, distance))

print("rotations", rotations[:5])
```

```python
current_pos = 50 #it is stated that the dial starts at 50
password = 0

def advance_what_direction(current_pos, direction, distance):
        if direction == 'R':
            new_pos = (current_pos + distance) % 100
        else:
            new_pos = (current_pos - distance) % 100
        return new_pos

for direction, distance in rotations:
    current_pos = advance_what_direction(current_pos, direction, distance)
    
    if current_pos == 0:
        password += 1
        
print (password)
```

```python
current_pos = 50
total_zero_count = 0

for direction, distance in rotations:
    if direction == 'R':
        total_zero_count += (current_pos + distance) // 100 - current_pos // 100
        current_pos = (current_pos + distance) % 100
    else: 
        current_pos -= 1
        total_zero_count += current_pos // 100 - (current_pos - distance) // 100
        current_pos = (current_pos + 1 - distance) % 100

print(total_zero_count)


```
